```python
%%writefile app.py

import streamlit as st
import numpy as np
import string
import pickle

st.set_option('deprecation.showfileUploaderEncoding',False) 

# On prends notre modèle dans le même répertoire
pipe = pickle.load(open('new_model.pkl','rb'))
mlb = pickle.load(open('mlb.pkl','rb'))

unpop_info = pickle.load(open('unpop_info.pkl','rb'))

# remet les colonnes impopulaires (les remets vides) pour pouvoir appliquer inverse_transform du MLB
# need 'tags_counts' et 'seuil'
def unPopBack(y, seuil):
    y_0 = np.zeros((y.shape[0], len(unpop_info['tags_counts'])), dtype=int)
    y_0[:,(unpop_info['tags_counts'] > unpop_info['seuil']).values.tolist()] = y
    return y_0

def main():
  st.markdown("<h1 style='text-align: center; color: White;background-color:#e84343'>Auto Tagger</h1>", unsafe_allow_html=True)
  st.markdown("<h3 style='text-align: center; color: Black;'>Posez votre question, en anglais</h3>", unsafe_allow_html=True)
  st.markdown("<h4 style='text-align: center; color: Black;'>Pour obtenir les tags proposés pour StackOverflow</h4>", unsafe_allow_html=True)
  # st.sidebar.header("What is this Project about?")
  # st.sidebar.text("It a Web app that would help the user in determining whether they will get admission in a Graduate Program or not.")
  # st.sidebar.header("What tools where used to make this?")
  # st.sidebar.text("The Model was made using a dataset from Kaggle along with using Kaggle notebooks to train the model. We made use of Sci-Kit learn in order to make our Linear Regression Model.")

  txt = st.text_area('Question à tagger', '''
     ''')

  texte = [txt]

  if st.button('HAL, Propose moi des Tags!'):
    Y_pred = pipe.predict(texte)
    res_mlb = unPopBack(Y_pred, 45)
    updated_res = mlb.inverse_transform(res_mlb)
    st.success('Tag(s) proposé(s): {}'.format(updated_res[0]))

if __name__ =='__main__':
  main()
```
